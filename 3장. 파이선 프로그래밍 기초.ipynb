> ## 1-3. 다중 할당과 swap

```python
# 여러 변수 한번에 할당
a, b, c = 10, 20, 30
print(f"a={a}, b={b}, c={c}")

# 같은 값을 여러 변수에 할당
x = y = z = 100
print(f"x={x}, y={y}, z={z}")

# 변수 값 교환 (swap)
first = "사과"
second = "바나나"
print(f"교환 전: first={first}, second={second}")

first, second = second, first  # 파이썬의 멋진 기능!
print(f"교환 후: first={first}, second={second}")
```

> ## 3-2. 리스트 메서드

```python
# 리스트에 요소 추가
scores = [85, 90, 78]
scores.append(95)  # 끝에 추가
print("append 후:", scores)

scores.insert(1, 88)  # 특정 위치에 삽입
print("insert 후:", scores)

scores.extend([100, 77])  # 여러 개 한번에 추가
print("extend 후:", scores)

# 리스트에서 요소 제거
scores.remove(77)  # 값으로 제거
print("remove 후:", scores)

popped = scores.pop()  # 마지막 요소 제거하고 반환
print(f"pop된 값: {popped}, 남은 리스트: {scores}")

scores.pop(0)  # 특정 인덱스 제거
print("첫 요소 pop 후:", scores)

# 리스트 정렬
random_nums = [45, 12, 78, 34, 90, 23]
random_nums.sort()  # 오름차순 정렬
print("오름차순:", random_nums)

random_nums.sort(reverse=True)  # 내림차순 정렬
print("내림차순:", random_nums)

# 리스트 정보
print(f"리스트 길이: {len(random_nums)}")
print(f"최대값: {max(random_nums)}")
print(f"최소값: {min(random_nums)}")
print(f"합계: {sum(random_nums)}")
print(f"평균: {sum(random_nums) / len(random_nums):.2f}")
```

> ## 3-3. 리스트 컴프리헨션 (고급)

```python
# 기본 방법
squares = []
for i in range(1, 6):
    squares.append(i ** 2)
print("제곱수 (기본):", squares)

# 리스트 컴프리헨션
squares = [i ** 2 for i in range(1, 6)]
print("제곱수 (컴프리헨션):", squares)

# 조건 포함
evens = [i for i in range(1, 11) if i % 2 == 0]
print("짝수만:", evens)

# 문자열 처리
words = ["apple", "banana", "cherry"]
upper_words = [w.upper() for w in words]
print("대문자 변환:", upper_words)

first_letters = [w[0] for w in words]
print("첫 글자만:", first_letters)
```

> ## 3-5. 딕셔너리 활용

```python
# 여러 사람 정보 관리
employees = [
    {"name": "김철수", "dept": "개발팀", "salary": 4500},
    {"name": "이영희", "dept": "마케팅", "salary": 3800},
    {"name": "박민수", "dept": "개발팀", "salary": 5200},
]

# 전체 직원 출력
for emp in employees:
    print(f"{emp['name']} ({emp['dept']}): {emp['salary']:,}만원")

# 개발팀만 필터링
dev_team = [emp for emp in employees if emp['dept'] == '개발팀']
print("\n개발팀:", dev_team)

# 급여 총합
total_salary = sum(emp['salary'] for emp in employees)
print(f"전체 급여: {total_salary:,}만원")

# 딕셔너리 컴프리헨션
name_to_salary = {emp['name']: emp['salary'] for emp in employees}
print("이름-급여 매핑:", name_to_salary)
```

> ## 4-4. while 반복문

```python
# 기본 while문
count = 0
while count < 5:
    print(f"카운트: {count}")
    count += 1

# 무한 루프와 break
print("\n숫자 맞추기 게임:")
import random
target = random.randint(1, 10)
attempts = 0

while True:
    attempts += 1
    guess = random.randint(1, 10)  # 실제로는 input() 사용
    print(f"추측: {guess}")

    if guess == target:
        print(f"정답! {attempts}번 만에 맞췄습니다.")
        break
    elif guess < target:
        print("더 큰 수입니다")
    else:
        print("더 작은 수입니다")

    if attempts >= 5:  # 최대 5번
        print("기회를 모두 사용했습니다")
        break
```

>## 4-5. 중첩 반복문과 제어문

> ## 5-2. 다양한 매개변수

```python
# 기본값 매개변수
def introduce(name, age=20, city="서울"):
    print(f"이름: {name}, 나이: {age}, 도시: {city}")

introduce("김철수")  # 기본값 사용
introduce("이영희", 25)  # age만 변경
introduce("박민수", 30, "부산")  # 모두 지정

# 가변 인자 (*args)
def sum_all(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_all(1, 2, 3))  # 6
print(sum_all(1, 2, 3, 4, 5))  # 15

# 키워드 가변 인자 (**kwargs)
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="홍길동", age=30, job="개발자")
```

> ## 5-3. 실용적인 함수 예제

```python
# 1. 계산기 함수
def calculator(a, b, operation="+"):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        return a / b if b != 0 else "0으로 나눌 수 없습니다"
    else:
        return "지원하지 않는 연산입니다"

print(calculator(10, 3, "+"))  # 13
print(calculator(10, 3, "-"))  # 7
print(calculator(10, 3, "*"))  # 30
print(calculator(10, 3, "/"))  # 3.333...

# 2. 데이터 검증 함수
def validate_email(email):
    if "@" not in email:
        return False
    if "." not in email.split("@")[1]:
        return False
    return True

emails = ["user@gmail.com", "invalid.email", "test@domain.co.kr"]
for email in emails:
    valid = validate_email(email)
    print(f"{email}: {'유효' if valid else '무효'}")

# 3. 통계 함수
def get_stats(numbers):
    if not numbers:
        return None

    stats = {
        "count": len(numbers),
        "sum": sum(numbers),
        "average": sum(numbers) / len(numbers),
        "max": max(numbers),
        "min": min(numbers)
    }
    return stats

data = [85, 92, 78, 95, 88]
result = get_stats(data)
print("\n통계 정보:")
for key, value in result.items():
    print(f"  {key}: {value}")
```

> ## 5-4. Lambda 함수

```python
# 람다 함수 기본
square = lambda x: x ** 2
print(f"5의 제곱: {square(5)}")

# 람다 함수 활용
numbers = [1, 2, 3, 4, 5]

# map 사용
squared = list(map(lambda x: x ** 2, numbers))
print(f"제곱: {squared}")

# filter 사용
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"짝수: {evens}")

# sorted에서 사용
students = [
    {"name": "김철수", "score": 85},
    {"name": "이영희", "score": 92},
    {"name": "박민수", "score": 78},
]

# 점수순 정렬
sorted_students = sorted(students, key=lambda x: x['score'], reverse=True)
print("\n성적순 정렬:")
for student in sorted_students:
    print(f"  {student['name']}: {student['score']}점")
```

> ## 6-3. 예외 처리

```python
# 기본 try-except
try:
    num = int(input("숫자를 입력하세요: "))
    result = 10 / num
    print(f"결과: {result}")
except ValueError:
    print("올바른 숫자를 입력하세요!")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다!")
except Exception as e:
    print(f"오류 발생: {e}")
finally:
    print("프로그램 종료")

# 실용적인 예외 처리
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None
    except TypeError:
        return "숫자만 입력 가능합니다"

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # None
print(safe_divide(10, "a")) # 숫자만 입력 가능합니다
```
