> ## 5-3. ì‹¤ìš©ì ì¸ í•¨ìˆ˜ ì˜ˆì œ

```python
# 1. ê³„ì‚°ê¸° í•¨ìˆ˜
def calculator(a, b, operation="+"):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        return a / b if b != 0 else "0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ìŠµë‹ˆë‹¤"
    else:
        return "ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°ì…ë‹ˆë‹¤"

print(calculator(10, 3, "+"))  # 13
print(calculator(10, 3, "-"))  # 7
print(calculator(10, 3, "*"))  # 30
print(calculator(10, 3, "/"))  # 3.333...

# 2. ë°ì´í„° ê²€ì¦ í•¨ìˆ˜
def validate_email(email):
    if "@" not in email:
        return False
    if "." not in email.split("@")[1]:
        return False
    return True

emails = ["user@gmail.com", "invalid.email", "test@domain.co.kr"]
for email in emails:
    valid = validate_email(email)
    print(f"{email}: {'ìœ íš¨' if valid else 'ë¬´íš¨'}")

# 3. í†µê³„ í•¨ìˆ˜
def get_stats(numbers):
    if not numbers:
        return None

    stats = {
        "count": len(numbers),
        "sum": sum(numbers),
        "average": sum(numbers) / len(numbers),
        "max": max(numbers),
        "min": min(numbers)
    }
    return stats

data = [85, 92, 78, 95, 88]
result = get_stats(data)
print("\ní†µê³„ ì •ë³´:")
for key, value in result.items():
    print(f"  {key}: {value}")
```
